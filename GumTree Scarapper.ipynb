```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.gumtree.sg/s-books-magazines/v1c18p1?ic=used'
data = requests.get(url)
```

```python
soup = BeautifulSoup(data.content, 'html.parser')
```

```python
count_total_ads = soup.find("p",{"class":"pagination-results-count"})
count_total_ads = count_total_ads.find("span").text
```

```python
count_total_ads
```

```python
count_total_ads = count_total_ads[:-4]
count_total_ads
```

```python
count_total_ads = count_total_ads[count_total_ads.rfind(" ")+1:]
count_total_ads
```

```python
import math
```

```python
total_pages = math.floor(int(count_total_ads) / 20)
total_pages
```

```python
MasterLinkList = list()
i = 1
base_url = 'https://www.gumtree.sg/s-books-magazines/page-{0}/v1c18p{1}?ic=used'
while i <= total_pages:
    MasterLinkList.append(base_url.format(i,i))
    i = i + 1
```

```python
class gumTreeBooks:
    def __init__(self,ad_title,ad_price,ad_date,ad_place,ad_condition,ad_description):
        self.ad_title = ad_title
        self.ad_price = ad_price
        self.ad_date = ad_date
        self.ad_place = ad_place
        self.ad_condition = ad_condition
        self.ad_description = ad_description
    
```

```python
from time import sleep
base_url = 'https://www.gumtree.sg'
SublinkList = list()
#try to iterate the page list and get the anchor tag
for masterUrl in MasterLinkList:
    data_oneurl=requests.get(masterUrl)
    sleep(5)
    soup = BeautifulSoup(data_oneurl.content,'html.parser')
    print("parsing url: " + masterUrl)
    div_resultView = soup.find("div", {"class": "view"})
    for link in div_resultView.find_all('a'):
        if link.get('href').startswith('/a-'):
            SublinkList.append(base_url + link.get('href'))
            print(base_url + link.get('href'))
    
```

```python
#read the values for each of the add and assign it to class 
FinalList = list()
for childUrl in SublinkList:
    data_child = requests.get(childUrl)
    sleep(5)
    soup = BeautifulSoup(data_child.content,'html.parser')
    print("Parsing URL: " + childUrl)
    ad_title = soup.find("span", {"class":"myAdTitle"}).text
    ad_price = soup.find("span", {"class":"amount"}).text
    #drill down for other variables
    ul_class = soup.find("ul", {"class":"selMenu"})
    dv_class = ul_class.find_all("span",{"class":"value"})
    ad_date = dv_class[0].text.strip()
    ad_place = dv_class[1].text.strip()
    ad_condition = dv_class[2].text.strip()
    ad_description = soup.find("div", {"class":"description"}).text
    FinalList.append(gumTreeBooks(ad_title,ad_price,ad_date,ad_place,ad_condition,ad_description))
```
